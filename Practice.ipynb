```python
h = ["kutta bhauka lal","kutta kaata hara","bhalu kaata lal"]
e = ["dog barked red","dog bit green","bear bit red"]
```

```python
l = h[0].split()
l.extend(h[1].split())
l.extend(h[2].split())
hv = list(set(l))
l = e[0].split()
l.extend(e[1].split())
l.extend(e[2].split())
ev = list(set(l))
```

```python
def init_prob():
    p = {}
    k = []
    for i in ev:
        for j in hv:
            k = k + [i + "|" + j]
    for i in range(len(k)):
        p[k[i]] = 1/(len(ev))
###print(k[i] + ":" + str(p[k[i]]))
```
